## Arrays

1D arrays

```python
import numpy as np
```

```python
x=np.array(range(4),dtype=float)
print(x)
```

```python
y=np.array([2.1,5,1.1,-2])
print(y)
```

arithmetic

```python
z=x+y
print(z)
```

```python
z=x*y
print(z)
```

```python
z=np.dot(x,y)
print(z)
```

```python
z=x**2
print(z)
```

manipulating arrays

```python
# procedure to exchange the first and third element
print(z)
z2=np.copy(z) # creates a copy of z.  NOTE: z2=z does not work
temp=z2[1]
z2[1]=z2[3]
z2[3]=temp
print(z2)
```

# 2D arrays

```python
m=np.zeros([5,6],float)  # makes a matrix (i.e. a two-dimensional array)
print(m)
```

```python
m2=np.array([[1,2],[2,3]],float) # makes array
m3=[[1,2],[2,3]] # makes a list (not an array)
print(m2)
print(m3)
```

```python
# rotation matrix
R=np.zeros([2,2],float)
thetaDeg=90
theta=np.pi*thetaDeg/180
R[0,0]=np.sin(theta);
R[1,1]=np.sin(theta);
R[0,1]=np.cos(theta);
R[1,0]=-np.cos(theta);
v=np.array([[1],[0]],float)
v2=np.dot(R,v)
print(v2)
```
